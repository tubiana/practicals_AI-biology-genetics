````python
#@title Get sequence from UNIPROT ID

#@markdown Please give the sequence id, separated by a coma, like `P1023,Q32034`

sequence_ids = "Q92889,P27697,P07992" #@param {type:"string"}

#@markdown > double click on the sequence to copy it.
import requests

sequence_ids_list = sequence_ids.split(',')

for sequence_id in sequence_ids_list:
  print(f"Sequence for {sequence_id}")
  response = requests.get(f'https://rest.uniprot.org/uniprotkb/{sequence_id}.fasta')
  sequence = "".join(response.text.split("\n")[1:])
  print(sequence)

````
